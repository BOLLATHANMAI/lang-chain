```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-pro",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
```

```python
help(ChatGoogleGenerativeAI)
```

```python
from langchain_core.messages import SystemMessage  ,HumanMessage
content = [
    SystemMessage(content = "give the code for the given problem in python"),
    HumanMessage(content = "palindrome"),
]
ai_msg = llm.invoke(content)
print(ai_msg.content)
```

```python
while True:
    question =  input("enter the question to ask!")
    if question == "exit":
        break
    content.append(HumanMessage(content=question))
    response = llm.invoke(content)
    content.append(SystemMessage(content=response.content))
    print("Ai" , response.content)
    print("........")
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
chain = llm | parser
chain.invoke("hello world")
```

```python
from langchain_core.prompts import ChatPromptTemplate
system = "transulate the following from english to {language}"
chat = ChatPromptTemplate((("system",system ), ("user","{text}")))
```

```python
chat.invoke({"language":"italy" , "text":"hello gemini"})
```

```python
chain = chat | llm | parser
```
